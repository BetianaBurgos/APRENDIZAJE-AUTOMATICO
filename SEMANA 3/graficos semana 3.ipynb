## CARGA DE DATOS

### En esta sección se importa el dataset de automóviles y se visualizan las primeras filas para comprender la estructura de los datos.

```python
import pandas as pd


df = pd.read_csv("Automobile.csv")


df.head()
```

# Gráfico de dispersión: mpg vs weight

### Este gráfico permite analizar la relación entre el peso del automóvil y su consumo de combustible. Se espera observar si existe alguna relación entre ambas variables.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df["weight"], df["mpg"], color="red")
plt.title("Relación entre peso y consumo de combustible")
plt.xlabel("Peso del automóvil")
plt.ylabel("Millas por galón (mpg)")
plt.grid()

plt.show()
```

### Se observa una relación inversa entre el peso del automóvil y el consumo de combustible. A medida que el peso aumenta, el rendimiento (mpg) disminuye. Esto indica que los autos más pesados consumen más combustible, siendo menos eficientes.

## HISTOGRAMA DE CONSUMO DE COMBUSTIBLE 

### El histograma permite visualizar la distribución de la variable mpg y analizar su comportamiento.

```python
plt.figure(figsize=(8,5))
plt.hist(df["mpg"], bins=15,color= 'red',edgecolor='black')
plt.title("Distribución del consumo de combustible")
plt.xlabel("Millas por galón (mpg)")
plt.ylabel("Frecuencia")

plt.show()
```

### La mayor cantidad de vehículos se concentra en valores intermedios de consumo. Se observan pocos valores extremos, lo que indica que la mayoría de los autos tienen un consumo moderado.

# Boxplot: cilindros por origen

## Este gráfico permite comparar la distribución del número de cilindros según el origen del automóvil.

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df["origin"], y=df["cylinders"],color='red')

plt.title("Distribución de cilindros por origen")
plt.xlabel("Origen del automóvil")
plt.ylabel("Número de cilindros")

plt.show()
```

### Se observan diferencias entre los orígenes:
### Los autos de Estados Unidos presentan mayor cantidad de cilindros y mayor variabilidad
### Europa y Japón tienen motores más pequeños y eficientes.

### Esto refleja distintos enfoques en el diseño automotor.

# Gráfico de barras: distribución por año y origen

## Este gráfico muestra cómo se distribuyen los automóviles según el año del modelo y su origen.

```python

df_grouped = df.groupby(["model_year", "origin"]).size().unstack()


df_grouped.plot(kind="bar", figsize=(10,6))

plt.title("Distribución de automóviles por año y origen")
plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de automóviles")

plt.show()
```

## Se observa que los vehículos estadounidenses predominan en los primeros años. Con el tiempo, aumenta la participación de automóviles europeos y japoneses, lo que indica una evolución del mercado hacia mayor diversidad y eficiencia.


## La visualización de datos permitió identificar patrones importantes:

### Existe una relación inversa entre peso y consumo.
### La mayoría de los autos tiene un consumo moderado.
### Hay diferencias claras entre los orígenes de los vehículos.
### El mercado automotor evolucionó con el tiempo.

### Estas herramientas son fundamentales en el aprendizaje automático, ya que permiten comprender mejor los datos antes de aplicar modelos predictivos.
